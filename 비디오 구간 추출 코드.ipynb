```python
import os
import xml.etree.ElementTree as ET
from moviepy.video.io.VideoFileClip import VideoFileClip

# XML 및 MP4 파일 폴더 경로
xml_folder_path = r"C:\Users\skyla\Downloads\238-2.실내(편의점, 매장) 사람 이상행동 데이터\01-1.정식개방데이터\Training\02.라벨링데이터\TL_03.이상행동_13.폭행"
mp4_folder_path = r"C:\Users\skyla\Downloads\238-2.실내(편의점, 매장) 사람 이상행동 데이터\01-1.정식개방데이터\Training\01.원천데이터\TS_03.이상행동_13.폭행"
output_folder_path = r"C:\Users\skyla\Downloads\238-2.실내(편의점, 매장) 사람 이상행동 데이터\01-1.정식개방데이터\Training\test"

# 폴더에서 파일 목록 가져오기
xml_files = [f for f in os.listdir(xml_folder_path) if f.endswith('.xml')]
mp4_files = [f for f in os.listdir(mp4_folder_path) if f.endswith('.mp4')]

# 매칭되는 XML과 MP4 파일 쌍 찾기
matched_files = []
for xml_file in xml_files:
    base_name = os.path.splitext(xml_file)[0]
    mp4_file = base_name + '.mp4'
    if mp4_file in mp4_files:
        matched_files.append((os.path.join(xml_folder_path, xml_file), os.path.join(mp4_folder_path, mp4_file)))

# fight_start 라벨을 가진 첫 번째 box frame 숫자 추출 함수
def get_first_fight_start_frame(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    for track in root.findall('track'):
        if track.get('label') == 'fight_start':
            first_box = track.find('box')
            if first_box is not None:
                return int(first_box.get('frame'))
    return None

# 비디오 세그먼트 추출 함수
def extract_video_segment(input_file, start_time, end_time, output_file):
    # 비디오 파일 열기
    video = VideoFileClip(input_file)
    
    # 시작 시간과 끝 시간에 해당하는 구간 추출
    video_segment = video.subclip(start_time, end_time)
    
    # 추출한 구간 저장
    video_segment.write_videofile(output_file, codec="libx264")

# 매칭된 파일 쌍 처리
output_index = 1
for xml_file, mp4_file in matched_files:
    first_frame = get_first_fight_start_frame(xml_file)
    
    if first_frame is not None:
        start_time = first_frame / 3
        end_time = start_time + 10
        output_video = os.path.join(output_folder_path, f"test{output_index}.mp4")
        
        # 비디오 세그먼트 추출
        extract_video_segment(mp4_file, start_time, end_time, output_video)
        print(f"Video segment from {start_time} to {end_time} seconds extracted to {output_video}")
        
        output_index += 1
    else:
        print(f"No 'fight_start' label found in {xml_file}.")
```
